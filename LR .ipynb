```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ecommerce Customers.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

#**Training Testing data**
###We plan to apply simple linear regression to predict Yearly Amount Spent using the information on Avg. Session length. For this purpose we use Yearly Amount Spent as our dependent variable.

```python
x=df[['Avg. Session Length']]
y=df['Yearly Amount Spent']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

##**Predicting the result**

```python
pred = lr.predict(x_test)
print(pred)
```

```python
plt.scatter(y_test,pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

##**MULTIPLE LINEAR REGRESSION**

```python
x2 = df.drop(['Email', 'Address', 'Avatar', 'Yearly Amount Spent'], axis=1)
y2 = df['Yearly Amount Spent']

x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, test_size=0.3, random_state=24)
```
