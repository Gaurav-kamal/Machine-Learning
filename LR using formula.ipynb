```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ecommerce Customers.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
x=df[['Avg. Session Length']]
y=df['Yearly Amount Spent']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
print(pred)
```

```python
plt.scatter(y_test,pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
m = lr.coef_
b = lr.intercept_
```

```python
new=[]
for i in x_test.values:
  y=m*(i)+b
  new.append(y)

print(new)
```

```python
pred3=pd.DataFrame(new)
pred3
```

```python
pred_new=pd.DataFrame(pred)
pred_new
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(pred_new, pred3))
print('MSE:', metrics.mean_squared_error(pred_new, pred3))
print('RMSE:', np.sqrt(metrics.mean_squared_error(pred_new, pred3)))
```
